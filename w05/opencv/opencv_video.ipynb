```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
cap = cv2.VideoCapture("people.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("no frame found")
        break
    
    cv2.imshow('video', frame)
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
def filter (image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blurred, 10, 70)
    ret, thresh = cv2.threshold(edged, 70, 255, cv2.THRESH_BINARY_INV)
    return thresh

cap = cv2.VideoCapture("people.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("no frame found")
        break
    
    cv2.imshow('filter', filter(frame))
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
