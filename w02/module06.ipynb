## Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('user_behavior_dataset.csv')
```

```python
data.head()
```

```python
# no none value detected (all purple)
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
data["Age"].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="Age", data=data)
```

```python
sns.scatterplot(data=data, x="Age", y="App Usage Time (min/day)")
```

```python
data.columns
```

```python
sns.lmplot(data=data, x="Age", y="App Usage Time (min/day)", hue="Gender", palette="muted")
```

```python
sns.scatterplot(data=data, x="Device Model", y="Number of Apps Installed")
```

```python
sns.jointplot(data=data, x="Age", y="Gender")
```

```python
sns.pairplot(data=data, hue="Gender")
```
